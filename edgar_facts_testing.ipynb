```python
def explore_dict(data):
    rows = []

    for account_key, account_values in data.items():
        for unit_key, unit_values in account_values.items():
            for item in unit_values:
                row = {"account_key": account_key, "unit_key": unit_key}
                row.update(item)
                rows.append(row)

    df = pd.DataFrame(rows)
    return df
```

```python
test_df = explore_dict(units_dict)
test2_df = test_df[test_df['accn'].isin(combined_accessionNumbers_for_filter)].dropna(subset=['frame'])
```

```python
test3_df = test2_df.drop(['accn', 'start', 'fy', 'fp', 'frame', 'filed', 'form'], axis=1)
test3_df['fact__unit'] = test3_df['account_key'] + '__' + test3_df['unit_key']
test3_df = test3_df.drop(['account_key', 'unit_key'], axis=1)
```

```python
pivot_df = test3_df.pivot_table(index='end', columns='fact__unit', values='val', aggfunc='first')
pivot_df.index = pd.to_datetime(pivot_df.index)
pivot_df_recent= pivot_df[pivot_df.index.isin(combined_accessionNumbers_for_filter.index)]
pivot_df_recent = pivot_df_recent.dropna(axis=1, how='all')
for col in pivot_df_recent.columns:
    if '_USD' in col or '_shares' in col and pivot_df_recent[col].abs().max() > 1_000:
        pivot_df_recent[col] = pivot_df_recent[col] / 1_000_000
        pivot_df_recent.rename(columns={col: col + '(millions)'}, inplace=True)
pivot_df_recent.name = ticker
pivot_df_recent
```
